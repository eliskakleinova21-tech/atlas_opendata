```python
import uproot
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt

# Open a file (change the filename to match what you downloaded)
#filename = "C:\Users\Eli≈°ka\OneDrive\Plocha\open data cern\4lep\4lep\Data\data_A.4lep.root"
filename = "4lep/Data/data_A.4lep.root"

# e.g. actual file name from ATLAS open data
file = uproot.open(filename)

# See what trees are available
file.keys()
```

```python
# Open the main tree (often called "mini" in ATLAS open data)
tree = file["mini"]

# List branches
tree.keys()
```

```python
# Example: read lepton transverse momentum (this branch name is just an example; 
# when you see actual branch names from tree.keys(), use one of those)
lepton_pt = tree["lep_pt"].array()  # or whatever the branch is actually called

# Make a simple histogram of the first lepton pt
first_lepton_pt = ak.to_numpy(lepton_pt[:, 0])  # first lepton in each event

plt.hist(first_lepton_pt, bins=50)
plt.xlabel("Leading lepton pT [GeV]")
plt.ylabel("Events")
plt.show()
```
